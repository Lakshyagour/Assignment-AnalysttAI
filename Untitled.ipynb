```python
import pdfplumber,re
import pandas as pd
names = []
mobiles = {}
offices = {}
residance = {}
emails = {}
office_address ={}
residance_address = {}
with pdfplumber.open("File.pdf") as pdf:
    for i in range(0,len(pdf.pages)):
        page = pdf.pages[i]
        page = page.crop((0,0,360,558))
        text = page.extract_text()
        lines = text.split("\n")
        name = ""
        for i in range(0,len(lines)):
            if re.search("^CA.", lines[i]) or re.search("^Shri", lines[i]):
                names.append(lines[i])
                name = lines[i]
                j=i+1
                text=[]
                while not(re.search("^CA.", lines[j]) or re.search("^Shri", lines[j])):
                    text.append(lines[j])
                    j+=1
                    if(j==len(lines)):
                        break
                for i in range(0,len(text)):
                    if re.search("^Mobile:", text[i]):
                        mobiles[name] = text[i][8:]
                    
                    
                    if re.search("^E-mail:", text[i]):
                        emails[name] = text[i][9:]
                        
                        
                    if re.search("^Office:", text[i]):
                        if re.search("^Office:.*91]", text[i]):
                            offices[name] = text[i][8:]
                        else:
                            oa = text[i][8:]
                            j = i+1
                            while not(re.search("^Residence:", text[j]) or re.search("^CA.", text[j]) or re.search("^Shri", text[j])):
                                oa+=text[j]
                                j+=1
                                if(j==len(text)):
                                    break
                            office_address[name] = oa
                            
                            
                    if re.search("^Residence:", text[i]):
                        if re.search("^Residence:.*91]", text[i]):
                            residance[name] = text[i][11:]
                        else:
                            ra = text[i][11:]
                            j = i+1
                            while not(re.search("^CA.", text[j]) or re.search("^Shri", text[j]) or re.search("denotes Country Code &", text[j])):
                                ra+=text[j]
                                j+=1
                                if(j==len(text)):
                                    break
                            residance_address[name] = ra
                            
df = pd.DataFrame()
df = df.append([mobiles,offices,residance,emails,office_address,residance_address],ignore_index=True)  
df = df.T
df = df.reset_index()
df = df.rename({'index':'Name',0:'Mobiles',1:'Office',2:'Residance',3:'Emails',4:'Office Address',5:'Residence Address'},axis=1)
df.to_csv('out.csv')
```
